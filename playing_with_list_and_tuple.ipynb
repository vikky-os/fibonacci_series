# mutable lists inside a tuple which will retain their id's but can be updated later

```python
"""#my own attempt to create a nested list then convert it into the tuple
and that is how a tuple is created with mutable obects in it and therfore
when we create such thing the inner lists might stay mutable but their id's
wont ever change as they are inside a tuple and this is how we created it"""
list1=[]
num=1
n=int(input("enter the number of items you want in your mutable nested tuple:"))
m=int(input("enter the number of items you want in each list:"))
for i in range(n):
  list1.append([])
  for j in range(m):
    list1[i].append(num)
    num+=1
print(list1)
print(tuple(list1))
```

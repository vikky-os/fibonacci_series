#RECURSION

```python
def fact(n):
    if n == 0:
        return 1
    return n * fact(n - 1)

n=int(input("enter a number for factorial:"))

fact(n)
```

# FIBONACCI SERIES

```python
def fibo(n):
    if n == 1 or n == 2:
        return 1
    return fibo(n - 1) + fibo(n - 2)

n=int(input("enter a number:"))

fibo(n)
```

## time taken to calculate the fibonacci of the number and counting the number of times the function has been recalled during execution of recursion

```python
import time
count=0
def fibo(n):
  global count
  count+=1
  if n == 1 or n == 2:
        return 1
  return fibo(n - 1) + fibo(n - 2)

n=int(input("num:"))
start = time.time()
fibo(n)
end = time.time()
print(f'It took approximately {round(end - start)} seconds.')

print(f'the number of times the fucntion has been recursed is: {count}')
print(f'the fibonacci value of {n} is {fibo(n)}')
```

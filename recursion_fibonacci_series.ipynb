#RECURSION

```python
def fact(n):
    if n == 0:
        return 1
    return n * fact(n - 1)

n=int(input("enter a number for factorial:"))

fact(n)
```

# FIBONACCI SERIES

```python
def fibo(n):
    if n == 1 or n == 2:
        return 1
    return fibo(n - 1) + fibo(n - 2)

n=int(input("enter a number:"))

fibo(n)
```

## time taken to calculate the fibonacci of the number

```python
import time
def fibo(n):
    if n == 1 or n == 2:
        return 1
    return fibo(n - 1) + fibo(n - 2)
n=int(input("num:"))
start = time.time()
fibo(n)
end = time.time()
print(f'It took approximately {round(end - start)} seconds.')
fibo(n)
```
